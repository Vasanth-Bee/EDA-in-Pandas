```python
df_Country = pd.read_excel('C:\Vasanth\Zomatodataset\Country-Code.xlsx')
```

```python
df_Country.head(5)
```

```python
final_df = pd.merge(df,df_Country,on ='Country Code',how ='left')
```

```python
final_df.head(2)
```

```python
country_val =final_df['Country'].value_counts().values
```

```python
country_name = final_df['Country'].value_counts().index
```

```python
## Top 3 Country
plt.pie(country_val[:3],labels= country_name[:3],autopct="%1.2f%%")
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head(5)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette =['white','red','orange','yellow','green','green'])
```

```python
### Find the contry name with 0 ratings
final_df[final_df["Rating color"] == 'White'].groupby("Country").size().reset_index()
```

```python
### Find out which currency is used by which country.

##final_df.columns

final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
## which country have online delivery options?

##final_df['Has Online delivery'].value_counts()

final_df[final_df['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index()
```

```python
##create a pie chart by city distribution
final_df['City'].value_counts().values
final_df['City'].value_counts().index

city_values = final_df['City'].value_counts().values
city_index = final_df['City'].value_counts().index

plt.pie(city_values[:5],labels =city_index[:5],autopct="%1.2f%%")
```
